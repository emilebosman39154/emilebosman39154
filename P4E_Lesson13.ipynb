### **Format Operator** (%)

**NB:**

1. In **integers** the % = *modulus operator*

2. In **strings** the % = the ***format operator*** (also called an ***interpolation operator***)





```python
camels = 52
horses = 2
print(camels)
'%d' % camels                               # 'd' stands for decimal AND removes the decimal
'I have spotted %d camels.' % horses        # adding the the second operand "% camels" takes the value of the variable camels and add it to the text
```

**Important:**

If you need to format MORE than one sequence in the string then the second argumet has to be a tuple.

Note:

**%d** = formatting an ***integer***

**%g** = formatting a ***float-point***

**%s** = formatting a ***string***

```python
'In %d years I have spotted %g %s.' % (3,0.1,'camels')      # (3,0.1,'camels') = a tuple    
```

Allways ensure the number of elements in the tuple corresponds to the number (and the correct type) of format sequences in the string.

Using a Dictionary as an argument:

The formats in the string must include a **parenthesised mapping key** into that **dictionary** inserted ***immediately after the '%' character***. The mapping key selects the value to be formatted from the mapping.

```python
print('%(language)s has %(number)03d quote types.' %      # 03 before the d spesifies there to be 3 characters (numbers)
      {'language': "Python", "number": 2})
```

### **Pseudo Code**

1. **Import** library named **random**

2. Announce your Generator to your user with the text:
**"Hello! Welcome to my Password Generator"**

3. **Create a variable named**: "**length**" that will
ask the user the following:
**"Enter the length of your required password"**
**PS.** Make sure this is created on a **NEW LINE** when the code runs (i.o.w. there must be a space between your "Hello, Welcome..."

4. **Create 4 variables:**
lower (for lowercase)

*   upper

5. **Combine these data** (of the above 4 variables)

6. Use random to **generate a RANDOM sample of the combined variables to the LENGTH specified** by the user

7. **Create the password**

8. **Show the password on screen**

**Password Generator Code:**

```python
# Import the required modules
import random
import string

print('Hello! Welcome to my Password Generator')

# Input the length of the Password as required
length = int(input('\nEnter the length of password required:  '))

# Define your data to be ued in the Password Generator
lower = "abcdefghijklmnopqrstuvwxyz"
upper = "ABCDEFGHIJKLMNOPQRSTUVWXYZ"
numbers = "0123456789"
symbols = "!@#$%^&*()<>'#:;"

# Alternative string module for above (hence the reason for import String Library is):
#lower = string.ascii_lowercase
#upper = string.ascii_uppercase
#number = string.digits
#symbols = string.puntuation 

# Combine the data
answer = lower + upper + numbers + symbols

# Use Random to create new Password each time
temp = random.sample(answer, length)

# Create the password
password = "".join(temp)

# Show the password on screen
print(password)
```

# **Chapter 7: Files**

**What are files?**

In essence. Well it really is a ***bunch of lines of characters***.

Sometimes it is created by the python code and sometimes it is a document that already exists (to which you apply some code).



**How:**

To read a file and work with it we use the **function open().**

open() returns a** file handle** which makes reading a file possible 

**Note: **this does not read the file it merely gets it ready for you to open it and work with it.

**Parameters:**

handle = open(filename,mode)
e.g. fhand = open('mbox.txt,'r') **r** of read and **w** for write

Above **fhand **is a variable and can be called anything (merely descriptive)

```python
fhand = open('mbox.txt')
print(fhand)

# 1. will return an error since it does not know the file (in its default directory)
# 2. it cannot find the file
# 3. Important - this will not open the text in the file (the handle is the 
# gateway to the file)  It will specify information about your file only.
```

**Reading Files:**

When we read through a text file Python does it reading line by line (thus the \n) as discused before.

```python
file = open('mbox.txt')
for text in file:      # read this as 'for each line in the file run this loop and then print'
  print(text)
```
