3. **Operations**

What is an Operator and and Operand

**Operator:** Operators in Python are used for operations between two **values** or **variables**

**Operand: **is what the operators are applied to (i.e. the values)

**Thus: **The operators indicate what action or operation to perform. The operands indicate what items to apply the action to



**Type Operators:**

7 Types (we only work now with Arithmetic Operators):

1. Arithmetic operators
2. Assignment operators 
3. Comparison operators
4. Logical operators
5. Membership operators
6. Identity operators
7. Bitwise operators


**Arithmetic Operators:**

```python
x = 10
y = 15
#addition
a = x + y
print (a)
#subtraction
b = x - y
print (b)
#multiplication
c = x * y
print (c)
#division
d = x / y
print (d)
#floor division
e = x // y
print (e)
#modulus
f = x % y
print (f)
#exponentiation
g = x ** y
print (g)
```
