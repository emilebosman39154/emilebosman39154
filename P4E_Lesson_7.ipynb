## **Basic Introduction to Turtle:**

Note: Run this code in PyCharm and not in Colab.

```python
# importing turtle package
import turtle

# The following code will draw a square
# starting with black as the default color
# then changing to red, blue and green

# set turtle shape
turtle.shape("turtle")
turtle.speed(2) #speed_value can be set to any integer from 1 (slowest) - 10 (fastest).

# set the colormode
# mode value can be between 0 - 255 (ideal: 255)
turtle.colormode(255)

# use forward by 100 pixel
# default pen color is black
turtle.forward(100)
# or turtle.backward()

# change the pencolor
# pencolor is red

turtle.pencolor("red")

# use forward by 100 pixel
# then 90 degrees right (or alternatively: turtle.left()
turtle.right(90)
turtle.forward(100)

# change the pencolor
# using r,g,b (reg, green, blue)
# pencolor is blue
turtle.pencolor((41, 41, 253))

# use forward by 100 pixel
# then 90 degrees right
turtle.right(90)
turtle.forward(100)

# change the pencolor
# pencolor is green
turtle.pencolor(41, 253, 41)

# use forward by 100 pixel
# then 90 degrees right
turtle.right(90)
turtle.forward(100)

turtle.done()

# Additional turtle codes
# 1. width()
# This makes the drawn line thicker, e.g
# e.g. width(15)

# 2. fillcolor()
# This is the filled color of a drawn object

# 3. Pen Up or Pen Down
# turtle.up() / turtle.penup() (will not leave a line on the screen)
# turtle.down() / turtle.pendown() (will draw a line on the screen)

# 4. goto()
# This moves the turtle to a position x,y

# 5. shape()
# This changes the shape to: 'arrow', 'classic'
# 'turtle' or 'circle'

# 6. Customise Screen Setup

# Set up screen ("screen" is the screen name given in this instance, but can be any name)
# turtle.setup(650,650)
# screen = turtle.Screen()
# screen.bgcolor('navy')
```

**Rainbow Benzene**



```python
# Python program to draw
# Rainbow Benzene
# using Turtle Programming
import turtle

screen = turtle.Screen()
colors = ['red', 'purple', 'blue', 'green', 'orange', 'yellow']
t = turtle.Pen()
turtle.bgcolor('black')
for x in range(360):
    t.pencolor(colors[x%6])  # Returns the remainder of the division by 6 - changes the pen colour
    t.width(x//100 + 1) # rounds to the nearest whole number - increase the width of the pen line slowly
    t.forward(x)
    t.left(59)

screen.exitonclick()
```

***Spirograph ***

```python
# Import the turtle library for
# drawing the required curve
import turtle as tt

# Set the background color as black,
# pensize as 2 and speed of drawing
# curve as 10(relative)
tt.bgcolor('black')
tt.pensize(2)
tt.speed(10)

# Iterate six times in total
for i in range(6):

    # Choose your color combination
    for color in ('red', 'magenta', 'blue',
                  'cyan', 'green', 'white',
                  'yellow'):
        tt.color(color)

        # Draw a circle of chosen size, 100 here
        tt.circle(100)

        # Move 10 pixels left to draw another circle
        tt.left(10)

    # Hide the cursor(or turtle) which drew the circle
    tt.hideturtle()

tt.exitonclick()
```

**Your First Turtle Game**

```python
import turtle

import turtle
import random
import time

screen = turtle.Screen()
screen.bgcolor("lightblue")

# Two players in game

# Player 1

player_1 = turtle.Turtle() #This creates your own Turtle
player_1.color("blue")
player_1.shape("turtle")

# Player 2

player_2 = player_1.clone()
player_2.color("red")

# Position Players
player_1.penup()
player_1.goto(-300,200)
player_2.penup()
player_2.goto(-300,-200)

# Draw a Finish Line
player_1.goto(300,-250)
player_1.left(90)
player_1.pendown()
player_1.color("black")
player_1.forward(500)
player_1.write("Finish!", font=75)
player_1.penup()
player_1.color("blue")
player_1.goto(-300,200)
player_1.right(90)

# Both players need pendown to see movement trail

player_1.pendown()
player_2.pendown()

# Creates values for the die

die = [1, 2, 3, 4, 5, 6]
# lets create the game

for i in range(30):
   if player_1.pos() >= (300,250):
       print("Player 1 wins the Race!") # This is printed in the console and not on screen
       break  # this is to break the for loop
   elif player_2.pos() >= (300,-250):
       print("Player 2 wins the Race!")
       break
   else:
       die_roll = random.choice(die)
       player_1.forward(30 * die_roll)  # increase the die_roll movement
       time.sleep(1.5) # 1.5 seconds delay
       die_roll2 = random.choice(die)
       player_2.forward(30 * die_roll2)  # increase the die_roll movement
       time.sleep(1.5)

screen.done()

# turtle.done() # this will keep turtle on the screen OR
# screen.exitonclick()
```
